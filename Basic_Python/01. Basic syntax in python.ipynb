#       PYTHON IDENTIFIER   

Python Identifiers are user-defined names to represent a variable, function, class, module or any other object.

If we assign some name to a programmable entity in Python, then it is nothing but technically called an identifier.

Python language lays down a set of rules for programmers to create meaningful identifiers.


#                             Guidelines For Creating Identifiers In Python.

1. To create an identifier, use a sequence of letters either in lowercase (a to z) or uppercase (A to Z).
   We can also mix up digits (0 to 9) or an underscore while writing an identifier.
   example ---> Names like shapeClass, shape_1, and upload_shape_to_db are all valid identifiers.
   
   
2. We cannot use digits to begin an identifier name. It will lead to syntax error.
   example –--> The name, 0Shape is incorrect, but shape1 is a valid identifier.
   
   
3. Python Identifiers cannot have special characters [‘.’, ‘!’, ‘@’, ‘#’, ‘$’, ‘%’] in their formation.

# LINES AND INDENTATION     

In python blocks of code are denoted by line indentation

All statements within the block must be indented the same space

In python all the continuous lines indented with same number of spaces forms a block.

In Python,indentation is required for indicating what block of code a statement belongs to.


```python
if True:                              #  The lines print("True") and print("False") are 2 separate code blocks.
    print("True")                     #  Here : dentes whatever comes ahead is the part of this block
else:
    print("False")
 
print("All Done")                    #    print("All done") is not indented, and so is not part of the else-block.
```

# MULTI-LINE STATEMENT 

Statements in Python typically end with a new line.

Python allows the use of line continuation character ---->   Back slash \ to denote that the line should continue

```python
item_1 = 200
item_2 = 400
item_3 = 600
```

```python
# But when we want to write the above 1 line statement on multiple lines we make use of backslash \
# Remove backslash and see what happens

total_item = item_1 + \
             item_2 + \
             item_3                          # Note -: if we give space after backslash our statement won't execute
```

```python
print(total_item)    # there is no need of single or double inverted comma inside print function statement
                     # as we are passing variables inside print function as a statement and not string
```

```python
# Statements contained within [], {}, or () brackets do not need line continuation character

days = ['Monday', 'Tuesday', 'Wednesday',
        'Thursday', 'Friday']
```

```python
print(days)
```

#   MULTIPLE STATEMENTS GROUP AS SUITE   

A group of individual statements,which make a single code block are called suites in Python.

Compound or complex statements, such as if, while, def, and class require a header line and a suite.

Header lines begin the statement (with the keyword) and terminate with a colon ( : ) and are followed by one or more lines

which make up the suite. For example −

```python
if expression :                 # here expression means our condition
   suite                        # code block(suite)
elif expression : 
   suite 
else : 
   suite
```

# Single Statement Suites

If the suite of an if clause consists only of a single line, it may go on the same line as the header statement.

```python
var = 100                                                  # single = means assign and double == means equal to 
if ( var == 100 ) : print("Value of expression is 100")
print("Good bye!")
```
